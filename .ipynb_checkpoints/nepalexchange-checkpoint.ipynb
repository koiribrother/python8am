```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('http://www.nepalstock.com/annualtrading/annual','html.parser')
soup = requests.get(url)
soup
```

```python
codes = soup.text
codes = BeautifulSoup(codes,"lxml")
codes
```

```python
table_codes = codes.table
tags = table_codes.find_all('tr')
tags
```

```python
data = []
for tag in tags:
    y = tag.text.split('\n')
    if y[1] !='':
        data.append(y[1:])
data
```

```python
import csv
with open('nepalexchange.csv','w') as f:
    x = csv.writer(f)
    x.writerows(data)
    
```
